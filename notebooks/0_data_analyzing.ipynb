# Data analyzing pipeline

This notebook aggregates the original data-analysis scripts for the China international citation ratio paper. It follows the lightweight research-notebook style of the reference repository while preserving the archived Python scripts as the source of truth.

The cells below call the original scripts as separate Python processes. This keeps each script's imports, globals, file writes, and working-directory assumptions isolated from the notebook environment. No scientific logic is reimplemented here.

## Execution convention

Run this notebook from the repository root or from the `notebooks/` folder. The helper below resolves the project root and runs analysis scripts with `code/` as the working directory. Under this convention, the original scripts write intermediate outputs under `code/p1_*`, `code/p3_*`, `code/p4_*`, `code/p5_*`, and `code/p6_*`, matching the plotting scripts' `..` path assumptions.

The original scripts expect large external input files under `/scratch/xy2949/OpenAlex_2025/pickles/`.

```python
from pathlib import Path
import subprocess
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

CODE_DIR = PROJECT_ROOT / "code"
ANALYSIS_DIR = CODE_DIR / "1_data_analyzing"

def run_archived_script(relative_script, working_dir=CODE_DIR):
    script_path = ANALYSIS_DIR / relative_script
    print(f"Running: {script_path.relative_to(PROJECT_ROOT)}")
    print(f"Working directory: {Path(working_dir).relative_to(PROJECT_ROOT)}")
    return subprocess.run([sys.executable, str(script_path)], cwd=str(working_dir), check=True)

print(f"Project root: {PROJECT_ROOT}")
```
