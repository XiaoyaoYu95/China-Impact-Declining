# Main figure plotting pipeline

This notebook aggregates the original figure-plotting scripts for the China international citation ratio paper. It provides a readable paper-oriented execution order while preserving the original plotting scripts as the source of truth.

Each code cell runs one archived plotting script as a separate Python process from `code/2_figure_plotting_main/`, preserving the scripts' relative-path behavior and output filenames.

## Execution helper

Run this notebook from the repository root or from `notebooks/`. The plotting scripts expect upstream result folders under `code/`, because they generally resolve `pre_path = os.path.abspath("..")` from `code/2_figure_plotting_main/`.

```python
from pathlib import Path
import subprocess
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

FIGURE_DIR = PROJECT_ROOT / "code" / "2_figure_plotting_main"

def run_plot_script(script_name, working_dir=FIGURE_DIR):
    script_path = FIGURE_DIR / script_name
    print(f"Running: {script_path.relative_to(PROJECT_ROOT)}")
    print(f"Working directory: {Path(working_dir).relative_to(PROJECT_ROOT)}")
    return subprocess.run([sys.executable, str(script_path)], cwd=str(working_dir), check=True)

print(f"Project root: {PROJECT_ROOT}")
```
